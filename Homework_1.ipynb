Read **Winston, p. 115, Problem 13**. Then complete the following code to solve the problem. You must use the outline given. Note that the variable names cannot be chosen arbitrarily; you will need to look at the full code outline to see what the variables should be named.

```python
from ortools.linear_solver import pywraplp

solver = pywraplp.Solver.CreateSolver("GLOP")

# Define the variables.
w1 = solver.NumVar(0,solver.infinity(), "wheat1")
w2 = solver.NumVar(0,solver.infinity(),"wheat2")
a1 = solver.NumVar(0,solver.infinity(),"alfalfa1")
a2 = solver.NumVar(0,solver.infinity(),"alfalfa2")

# Set the objective. Explain how you obtained it in the text box after the code block.
solver.Maximize(1*(w1+a1)+0.8*w2+0.9*a2)

# Add the constraints.
solver.Add(w1 >= 4*a1) #percentage of wheat in Feed 1
solver.Add(a2 >= 1.5*w2) #percentage of alfalfa in Feed 2
solver.Add(w1+w2 <= 1000) #wheat
solver.Add(a1+a2 <= 800) #alfalfa

# Solve and print the solution. DO NOT CHANGE ANYTHING PAST THIS POINT.
status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
  print("Solution:")
  print(f"Optimal value = {solver.Objective().Value():0.2f}")
  print(f"{w1} = {w1.solution_value():0.2f}")
  print(f"{w2} = {w2.solution_value():0.2f}")
  print(f"{a1} = {a1.solution_value():0.2f}")
  print(f"{a2} = {a2.solution_value():0.2f}")
else:
  print("The problem does not have an optimal solution.")
```

# Problem 5

Read **Winston, p. 120, Problem 46**. Then complete the following code to solve the problem as a linear program.

```python
solver = pywraplp.Solver.CreateSolver("GLOP")

# Complete the definitions of the variables. Use NumVar, not IntVar.
f12 = solver.NumVar(0, solver.infinity(), "f12") #full time 12-1
f1 = solver.NumVar(0, solver.infinity(), "f1") #full time 1-2
p9 = solver.NumVar(0, solver.infinity(), "p9") #part time 9-12
p10 = solver.NumVar(0, solver.infinity(), "p10") #part time 10-1
p11 = solver.NumVar(0, solver.infinity(), "p11") #part time 11-2
p12 = solver.NumVar(0, solver.infinity(), "p12") #part time 12-3
p1 = solver.NumVar(0, solver.infinity(), "p1") #part time 1-4
p2 = solver.NumVar(0, solver.infinity(), "p2") #part time 2-5

# Set the objective
solver.Minimize(8*8*(f12+f1) + 3*5*(p9 + p10 + p11 + p12 + p1 + p2))

# Add the constraints
solver.Add(f12 + f1 + p9 + p10 + p11 >= 4) #9-10 required tellers
solver.Add(f12 + f1 + p9 + p10 + p11 + p12 >= 3) #10-11 required tellers
solver.Add(f12 + f1 + p9 + p10 + p11 + p12 >= 4) #11-12 required tellers
solver.Add(f1 + p9 + p10 + p11 + p12 >= 6) #12-1 required tellers
solver.Add(f12 + p10 + p11 + p12 + p1 >= 5) #1-2 required tellers
solver.Add(f12 + p11 + p12 + p1 + p2 >= 6) #2-3 required tellers
solver.Add(f12 + f1 + p12 + p1 + p2 >= 8) #3-4 required tellers
solver.Add(f12 + f1 + p1 + p2 >= 8) #4-5 required tellers
solver.Add(p9 + p10 + p11 + p12 + p1 + p2 <=5) #limit in part time tellers

# Solve and print the solution. DO NOT CHANGE ANYTHING PAST THIS POINT.
status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
  print("Solution:")
  print(f"Optimal value = {solver.Objective().Value():0.2f}")
  print(f"{f12} = {f12.solution_value():0.2f}")
  print(f"{f1} = {f1.solution_value():0.2f}")
  print(f"{p9} = {p9.solution_value():0.2f}")
  print(f"{p10} = {p10.solution_value():0.2f}")
  print(f"{p11} = {p11.solution_value():0.2f}")
  print(f"{p12} = {p12.solution_value():0.2f}")
  print(f"{p1} = {p1.solution_value():0.2f}")
  print(f"{p2} = {p2.solution_value():0.2f}")
else:
  print("The problem does not have an optimal solution.")
```
